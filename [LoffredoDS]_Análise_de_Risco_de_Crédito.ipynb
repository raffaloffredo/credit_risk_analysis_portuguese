# Sobre o Projeto

A anlise de crdito nas instituies financeiras  crucial para avaliar se um tomador de emprstimo tem potencial para cumprir com o contrato ou se ele poder se tornar inadimplente, ou seja, deixar de pagar o emprstimo efetuado. Isso porque emprestar ou no ir impactar diretamente o balano da instituio financeira, podendo alavancar os resultados ou prejudic-lo consideravelmente.

<br/>

Essa avaliao do crdito geralmente envolve analisar o histrico de crdito do cliente, como por exemplo se ele j foi inadimplente anteriormente ou no, se ele possui bens que poderiam servir como garantia, entre diversos outros fatores que visam medir a capacidade dele em cumprir com suas obrigaes de devolver o valor concedido.

<br/>

Quando o cliente se torna inadimplente, ou seja, deixa de pagar sua dvida, essa pessoa se torna um _default_. E, em razo do risco de default tanto paras as instituies financeiras, quanto para todo o sistema financeiro e at mesmo para a economia de um pas, cada vez mais se investe em modelos de inteligncia artificial para prever os possveis defaults e minimiz-los, com a finalidade de evitar um prejuzo maior.

## 4. Funo `val_model`

Para iniciar essa etapa, em primeiro lugar vou construir uma funo denominada `val_model` com a construo de um *pipeline* (fluxo de trabalho) que ir padronizar os dados com o  *Standard Scaler* (uma vez que temos presena de outliers no dataset), classific-los (de acordo com o algoritmo que for inserido na funo pelo parmetro `clf`) e fazer a avaliao do modelo por meio da validao cruzada, levando em considerao o valor de Recall obtido. O resultado final dessa funo ser mostrar o valor mdio de Recall encontrado na validao cruzada.

```python
# construir funo de avaliao de modelos
def val_model(X, y, clf, quite=False):
    """
    Realiza cross-validation com os dados de treino para um determinado modelo.

    # Arguments
        X: Data Frame, contm as variveis independentes.
        y: Series, vetor contendo a varivel alvo.
        clf: modelo classificador do scikit-learn.
        quite: bool, indicando se a funo deve imprimir os resultados ou no.

    # Returns
        float, mdia dos scores da cross-validation.
    """

    # converter variveis em arrays
    X = np.array(X)
    y = np.array(y)

    # criao de fluxo de trabalho
    ## 1. padronizar os dados com o StandardScaler
    ## 2. classificar os dados
    pipeline = make_pipeline(StandardScaler(), clf)

    # avaliao do modelo por validao cruzada
    ## de acordo com o valor de Recall
    scores = cross_val_score(pipeline, X, y, scoring='recall')

    # mostrar mdia do valor de Recall e desvio padro do modelo
    if quite == False:
        print("Recall: {:.4f} (+/- {:.2f})".format(scores.mean(), scores.std()))

    # retornar a mdia dos valores de Recall obtidos na validao cruzada
    return scores.mean()
```

# Mtricas de Avaliao de Performance

Para avaliar o desempenho dos modelos a mtrica que ser utilizada  o valor de Recall. Isso porque quando se tem dados desbalanceados, como  o caso do conjunto em estudo, ainda que ele j tenha sido tratado e balanceado, a acurcia no  uma boa mtrica de avaliao. O motivo disso se deve ao fato de que se pode obter um resultado de acurcia muito alto, entretanto, na identificao de default, a deteco ter um resultado muito baixo, o que no nos leva ao nosso objetivo.

<br/>

O Recall  a mtrica que nos fornece a melhor medida para o problema especfico em estudo. Isso se deve porque, no caso de default, os Falsos Negativos so mais prejudiciais para uma empresa do que os Falsos Positivos. Em outras palavras,  melhor que o modelo erre ao dizer que um cliente  default, quando na realidade ele no seja. Do que erre ao apontar um cliente como no sendo default, quando na verdade ele , trazendo prejuzos para os negcios.

<br/>

Tendo isso em vista, busca-se uma alta taxa de Recall.

O Recall olha para todas os defaults e visa responder: o quanto o modelo acerta? O resultado ser um valor entre 0 e 1, sendo que, quanto mais prximo de 1, melhor, pois indica uma baixa taxa de Falsos Negativos.

<br/>

Seu clculo  dado por:

<br/>

$${\text{Recall: }}\frac{\text{Verdadeiro Positivo}} {\text{Verdadeiro Positivo + Falso Negativo}}$$

<br/>

Ainda pensando na finalidade desse estudo, outra mtrica de avaliao de modelos de classificao que pode ser utilizada  a AUC - *Area Under the Curve*, traduzindo  a Area abaixo da curva. Ela deriva do ROC - *Receiver Operating Characteristic* - e indica para ns o quanto o modelo consegue distinguir entre duas coisas. No nosso exemplo, entre um cliente default e no-default. O valor dado por AUC varia entre 0 e 1, sendo que, quanto mais prximo de 1, indica que melhor  o modelo.

<br/>

Por fim, a matriz de confuso compara os valores previstos com os valores reais, mostrando os erros e acertos do modelo. Sua sada tem 4 valores diferentes, dispostos da seguinte forma:

Em razo da quantidade de hiperparmetros que precisam ser ajustados,  necessrio encontrar o melhor valor para cada um deles em processos distintos. Para iniciar, se definiu a taxa de aprendizado em 0.1. A recomendao  de seja entre 0.05 e 0.3, de acordo com o problema a ser resolvido.

Vamos fazer a busca para o melhor parmetro para `n_estimators` que  a quantidade de rvores de deciso que deve conter no algoritmo. E, vamos utilizar um valor de semente para poder reproduzir os resultados.

```python
# definir uma semente para reprodutibilidade
seed = 123

# instanciar o modelo XGBoost
# definir a taxa de aprendizado em 0.1 e definir a semente
xgb = XGBClassifier(learning_rate=0.1, random_state=seed)

# definir dicionrio para descobrir a quantidade ideal de rvores
# em um range de 0 a 500 com incremento de 50
param_grid = {'n_estimators':range(0,500,50)}

# configurar validao cruzada com 10 dobras estratificadas
# shuffle=True para embaralhar os dados antes de dividir e definir a semente
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)

# configurao da busca de combinaes cruzadas com o classificador XGBoost
# parmetros de rvores do param_grid
# scoring: mtodo de avaliao por Recall
# n_jobs=1 para busca em paralelo (usando todos os ncleos disponveis)
# cv: estratgia de validao
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca do melhor parmetro para 'n_estimators'
grid_result = grid_search.fit(X_train_rus, y_train_rus)

# imprimir o melhor parmetro encontrado para 'n_estimators'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

O melhor resultado encontrado foi com 100 rvores de deciso e, com apenas esse hiperparmetro j superamos o _LGBMClassifier_	que ficou em primeiro lugar na lista com um Recall de 0.656231.

Com esse parmetro, como o incremento definido foi de 50, podemos refinar a busca em um intervalo menor e com passos menores:

```python
# definir uma semente para reprodutibilidade
seed = 123

# instanciar o modelo XGBoost
# definir a taxa de aprendizado em 0.1 e definir a semente
xgb = XGBClassifier(learning_rate=0.1, random_state=seed)

# definir dicionrio para descobrir a quantidade ideal de rvores
# em um range de 25 a 125 com incremento de 5
param_grid = {'n_estimators':range(25,125,5)}

# configurar validao cruzada com 10 dobras estratificadas
# shuffle=True para embaralhar os dados antes de dividir e definir a semente
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)

# configurao da busca de combinaes cruzadas com o classificador XGBoost
# parmetros de rvores do param_grid
# scoring: mtodo de avaliao por Recall
# n_jobs=1 para busca em paralelo (usando todos os ncleos disponveis)
# cv: estratgia de validao
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca do melhor parmetro para 'n_estimators'
grid_result = grid_search.fit(X_train_rus, y_train_rus)

# imprimir o melhor parmetro encontrado para 'n_estimators'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

Com essa nova busca, o melhor valor encontrado para a quantidade de rvores foi de 85.

Com o melhor parmetro para `n_estimators`, podemos defini-lo na atribuio do classificador e prosseguir para determinar os melhores valores para `max_depth`, que determina a profundidade das rvores de deciso e, para o `min_child_weight`, ou seja, o peso mnimo do n para que seja criado um novo n.

```python
# definir uma semente para reprodutibilidade
seed = 123

# instanciar o modelo XGBoost
# definir a taxa de aprendizado em 0.1 e definir a semente
## n_estimators: 85
xgb = XGBClassifier(learning_rate=0.1, n_estimators=85, random_state=seed)

# definir dicionrio para descobrir:
## a profundidade ideal em um range de 1 a 7 com incremento de 1
## o peso mnimo do n em um range de 1 a 4 com incremento de 1
param_grid = {'max_depth':range(1,8,1),
              'min_child_weight':range(1,5,1)}

# configurar validao cruzada com 10 dobras estratificadas
# shuffle=True para embaralhar os dados antes de dividir e definir a semente
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)

# configurao da busca de combinaes cruzadas com o classificador XGBoost
# parmetros de rvores do param_grid
# scoring: mtodo de avaliao por Recall
# n_jobs=1 para busca em paralelo (usando todos os ncleos disponveis)
# cv: estratgia de validao
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca do melhor parmetro para 'max_depth' e 'min_child_weight'
grid_result = grid_search.fit(X_train_rus, y_train_rus)

# imprimir o melhor parmetro encontrado para 'max_depth' e 'min_child_weight'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

Os valores obtidos para a profundidade (`max_depth`) foi de 6 e o `min_child_weight`  de 1. Temos mais 2 parmetros para ajustar!

Podemos inserir os novos parmetros encontrados e partir para encontrar o prximo: `gamma`. Esse parmetro define a complexidade das rvores no modelo.

```python
# definir uma semente para reprodutibilidade
seed = 123

# instanciar o modelo XGBoost
# definir a taxa de aprendizado em 0.1 e definir a semente
## n_estimators: 85
## max_depth: 6
## min_child_weight: 1
xgb = XGBClassifier(learning_rate=0.1, n_estimators=85, max_depth=6, min_child_weight=1, random_state=seed)

# definir dicionrio para descobrir:
## a profundidade ideal em um range de 0.0 a 0.4
param_grid = {'gamma':[i/10.0 for i in range(0,5)]}

# configurar validao cruzada com 10 dobras estratificadas
# shuffle=True para embaralhar os dados antes de dividir e definir a semente
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)

# configurao da busca de combinaes cruzadas com o classificador XGBoost
# parmetros de rvores do param_grid
# scoring: mtodo de avaliao por Recall
# n_jobs=1 para busca em paralelo (usando todos os ncleos disponveis)
# cv: estratgia de validao
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca do melhor parmetro para 'gamma'
grid_result = grid_search.fit(X_train_rus, y_train_rus)

# imprimir o melhor parmetro encontrado para 'gamma'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

Seguimos inserindo esses valores na atribuio do modelo de classificao e vamos buscar agora o melhor valor para a taxa de aprendizado (`learning_rate`).

```python
# definir uma semente para reprodutibilidade
seed = 123

# instanciar o modelo XGBoost
# definir a semente
## n_estimators: 85
## max_depth: 6
## min_child_weight: 1
## gamma = 0.1
xgb = XGBClassifier(n_estimators=85, max_depth=6, min_child_weight=1, gamma=0.1, random_state=seed)

# definir dicionrio para descobrir a taxa de aprendizado ideal
param_grid = {'learning_rate':[0.01, 0.05, 0.1, 0.2]}

# configurar validao cruzada com 10 dobras estratificadas
# shuffle=True para embaralhar os dados antes de dividir e definir a semente
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)

# configurao da busca de combinaes cruzadas com o classificador XGBoost
# parmetros de rvores do param_grid
# scoring: mtodo de avaliao por Recall
# n_jobs=1 para busca em paralelo (usando todos os ncleos disponveis)
# cv: estratgia de validao
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca do melhor parmetro para 'learning_rate'
grid_result = grid_search.fit(X_train_rus, y_train_rus)

# imprimir o melhor parmetro encontrado para 'learning_rate'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

# XGBoost: Engenharia de Atributos + Otimizao de Hiperparmetros

Aqui, vamos repetir o processo de otimizao dos hiperparmetros do algoritmo XGBoost para encontrar os valores que produzem os melhores desempenhos.

Vou comear a busca para o melhor valor para `n_estimators` que  a quantidade de rvores de deciso que deve conter no algoritmo.

```python
# definir uma semente para reprodutibilidade
seed = 123

# instanciar o modelo XGBoost
# definir a taxa de aprendizado em 0.1 e definir a semente
xgb = XGBClassifier(learning_rate=0.1, random_state=seed)

# definir dicionrio para descobrir a quantidade ideal de rvores
# em um range de 0 a 500 com incremento de 50
param_grid = {'n_estimators':range(0,500,50)}

# configurar validao cruzada com 10 dobras estratificadas
# shuffle=True para embaralhar os dados antes de dividir e definir a semente
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)

# configurao da busca de combinaes cruzadas com o classificador XGBoost
# parmetros de rvores do param_grid
# scoring: mtodo de avaliao por Recall
# n_jobs=1 para busca em paralelo (usando todos os ncleos disponveis)
# cv: estratgia de validao
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca do melhor parmetro para 'n_estimators'
grid_result = grid_search.fit(X_fe_train_rus, y_fe_train_rus)

# imprimir o melhor parmetro encontrado para 'n_estimators'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

O melhor resultado encontrado foi com 50 rvores de deciso. Com esse parmetro, como o incremento definido foi de 50, podemos refinar a buscar em torno desse valor, com passos menores:

```python
# definir uma semente para reprodutibilidade
seed = 123

# instanciar o modelo XGBoost
# definir a taxa de aprendizado em 0.1 e definir a semente
xgb = XGBClassifier(learning_rate=0.1, random_state=seed)

# definir dicionrio para descobrir a quantidade ideal de rvores
# em um range de 25 a 75 com incremento de 5
param_grid = {'n_estimators':range(25,75,5)}

# configurar validao cruzada com 10 dobras estratificadas
# shuffle=True para embaralhar os dados antes de dividir e definir a semente
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)

# configurao da busca de combinaes cruzadas com o classificador XGBoost
# parmetros de rvores do param_grid
# scoring: mtodo de avaliao por Recall
# n_jobs=1 para busca em paralelo (usando todos os ncleos disponveis)
# cv: estratgia de validao
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca do melhor parmetro para 'n_estimators'
grid_result = grid_search.fit(X_fe_train_rus, y_fe_train_rus)

# imprimir o melhor parmetro encontrado para 'n_estimators'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

Com essa nova busca, o melhor valor encontrado para a quantidade de rvores foi de 25. Dessa vez, apenas com esse ajuste melhoramos o desempenho do algoritmo de 0.642724 para 0.6620. Esse resultado j garantiria o segundo lugar nos modelos base!

Temos, ainda, outros 4 *tunnings* para ajustar, ento, vamos prosseguir.

Com o melhor parmetro para `n_estimators`, podemos defini-lo na atribuio do classificador e prosseguir para determinar os melhores valores para `max_depth`, que determina a profundidade das rvores de deciso e, para `min_child_weight`, ou seja, o peso mnimo do n para que seja criado um novo n.

```python
# definir uma semente para reprodutibilidade
seed = 123

# instanciar o modelo XGBoost
# definir a taxa de aprendizado em 0.1 e definir a semente
## n_estimators: 25
xgb = XGBClassifier(learning_rate=0.1, n_estimators=25, random_state=seed)

# definir dicionrio para descobrir:
## a profundidade ideal em um range de 1 a 7 com incremento de 1
## o peso mnimo do n em um range de 1 a 4 com incremento de 1
param_grid = {'max_depth':range(1,8,1),
              'min_child_weight':range(1,5,1)}

# configurar validao cruzada com 10 dobras estratificadas
# shuffle=True para embaralhar os dados antes de dividir e definir a semente
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)

# configurao da busca de combinaes cruzadas com o classificador XGBoost
# parmetros de rvores do param_grid
# scoring: mtodo de avaliao por Recall
# n_jobs=1 para busca em paralelo (usando todos os ncleos disponveis)
# cv: estratgia de validao
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca do melhor parmetro para 'max_depth' e 'min_child_weight'
grid_result = grid_search.fit(X_fe_train_rus, y_fe_train_rus)

# imprimir o melhor parmetro encontrado para 'max_depth' e 'min_child_weight'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

Podemos inserir os novos parmetros encontrados e partir para encontrar o prximo: `gamma`. Esse parmetro define a complexidade das rvores no modelo.

```python
# definir uma semente para reprodutibilidade
seed = 123

# instanciar o modelo XGBoost
# definir a taxa de aprendizado em 0.1 e definir a semente
## n_estimators: 25
## max_depth: 6
## min_child_weight: 1
xgb = XGBClassifier(learning_rate=0.1, n_estimators=25, max_depth=6, min_child_weight=2, random_state=seed)

# definir dicionrio para descobrir:
## a profundidade ideal em um range de 0.0 a 0.4
param_grid = {'gamma':[i/10.0 for i in range(0,5)]}

# configurar validao cruzada com 10 dobras estratificadas
# shuffle=True para embaralhar os dados antes de dividir e definir a semente
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)

# configurao da busca de combinaes cruzadas com o classificador XGBoost
# parmetros de rvores do param_grid
# scoring: mtodo de avaliao por Recall
# n_jobs=1 para busca em paralelo (usando todos os ncleos disponveis)
# cv: estratgia de validao
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca do melhor parmetro para 'gamma'
grid_result = grid_search.fit(X_fe_train_rus, y_fe_train_rus)

# imprimir o melhor parmetro encontrado para 'gamma'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

Mais uma vez, o melhor valor encontrado foi o valor mnimo, ento, prosseguimos inserindo esse valor na atribuio do modelo de classificao e buscando o melhor valor para a taxa de aprendizado (`learning_rate`).

```python
# definir uma semente para reprodutibilidade
seed = 123

# instanciar o modelo XGBoost
# definir a semente
## n_estimators: 25
## max_depth: 6
## min_child_weight: 1
## gamma = 0.1
xgb = XGBClassifier(n_estimators=25, max_depth=6, min_child_weight=2, gamma=0.0, random_state=seed)

# definir dicionrio para descobrir a taxa de aprendizado ideal
param_grid = {'learning_rate':[0.01, 0.05, 0.1, 0.2]}

# configurar validao cruzada com 10 dobras estratificadas
# shuffle=True para embaralhar os dados antes de dividir e definir a semente
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)

# configurao da busca de combinaes cruzadas com o classificador XGBoost
# parmetros de rvores do param_grid
# scoring: mtodo de avaliao por Recall
# n_jobs=1 para busca em paralelo (usando todos os ncleos disponveis)
# cv: estratgia de validao
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca do melhor parmetro para 'learning_rate'
grid_result = grid_search.fit(X_fe_train_rus, y_fe_train_rus)

# imprimir o melhor parmetro encontrado para 'learning_rate'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

# Teste de Hiptese

Apesar de numericamente observarmos que o Recall do modelo com a engenharia de atributos ter sido 3.25% superior ao modelo sem a engenharia, podemos realizar um teste estatstico e comprovar se esse melhor desempenho  significativo ou no.

<br/>

Para isso, vou utilizar um teste de propores conhecido como teste Z, para duas propores, que  o que melhor se adequa ao nosso caso. Ou seja, ele determina se as diferenas entre as propores de duas amostras independentes so estatisticamente significativas. Para isso, vou considerar as seguintes hipteses:

<br/>

* Hiptese Nula ($H_0$)  $\text{p-value}>$ 0.05

  **No h diferena significativa** no valor de Recall entre o modelo com e sem engenharia de atributos.

  <br>

* Hiptese Alternativa ($H_1$)  $\text{p-value} <=$  0.05

  Rejeita a hiptese nula, ou seja, **existe diferena** entre os dois modelos avaliados.

<br/>

_Nota_

O nvel de significncia (*p-value*)  um valor determinado de acordo com cada caso para determinar um limite para aceitar ou rejeitar a hiptese nula. Assim, quanto menor for o valor dado por *p-value*, mais forte so as evidncias contra a hiptese nula. or exemplo, se o resultado obtido em p-value fosse de 0.04, logo, conseguiramos rejeitar a hiptese nula e defender que o modelo com engenharia de atributos  superior com um nvel de confiana de 96%.

<br/>

Portanto, se o valor ficar acima do threshold, no conseguiremos rejeitar a hiptese nula, mas isso no  uma verdade absoluta. Significa apenas que no encontramos evidncias suficientes para rejeit-la.


```python
# definir variveis com os valores necessrios para o teste estatstico
## valor do recall do modelo sem engenharia de atributos
recall_model    = 0.6547
## valor do recall do modelo com engenharia de atributos
recall_model_fe = 0.6872
## quantidade de clientes que foram avaliados
n_clients = X_test.shape[0]

# quantidade de acertos de cada modelo
model_success    = int(recall_model * n_clients)
model_fe_success = int(recall_model_fe * n_clients)

# criar funo para calcular o teste z para duas propores
def z_test(success_a, success_b, n):
    # calcular parmetros
    p1 = success_a / n
    p2 = success_b / n
    p = (success_a + success_b) / (n + n)
    # definir frmula com o threshold de p = 0.05
    z_stat = (p1 - p2) / ((p * (1 - p) * (1/n + 1/n))**0.5)

    # calcular o valor-p
    p_val = (1 - norm.cdf(abs(z_stat))) * 2
    return z_stat, p_val

# aplicar a funo aos modelos de machine-learning
z_stat, p_val = z_test(model_success, model_fe_success, n_clients)
print(z_stat, p_val)
```

O valor de p-value do teste estatstico realizado foi igual a 4.41e-08, logo o valor de p encontrado  menor do que o valor de nvel de significncia, de 0.05. Com isso podemos rejeitar a hiptese nula e aceita-se que existe diferena entre os dois modelos de machine-learning criados. E, por isso, podemos dizer com suporte estatstico que o modelo com engenharia de atributos  superior ao modelo sem a engenharia.
